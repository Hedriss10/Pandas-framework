```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_csv('../base/imoveis_residencial.csv', sep=';')
```

```python
dados.head(5)
```

### Selecione somente os imóveis classificados com o tipo de "Apartamento"

```python
selecao = dados["Tipo"] == "Apartamento"
selecao
```

### Atribuindo dentro do banco de dados

```python
baseOne = dados[selecao]
baseOne
```

```python
print(f'Os Registros de apartamentos é {baseOne.shape[0]}')
```

#### Selecione somente os imóveis classificados com o tipos 'Casa', 'Casa de condomínio' e 'Casa da vila'

<p>Atenção para o macete, sempre é importante agrupar os dados de comparação para que o pandas possa gerar e executar</p>

```python
questionselecao = (dados["Tipo"] == 'Casa') | (dados["Tipo"] == 'Casa de Condomínio') | (dados["Tipo"] == 'Casa de Vila')
questionselecao
```

```python
baseTwo = dados[questionselecao]
baseTwo
```

```python
print(f'Os registros de casa, condomínio e vila é {baseTwo.shape[0]}')
```

Selecione somente os imóveis com as área entre 60 e 100 metros quadrados, incluindo os limites

```python
selecao = (dados["Area"] >= 60) & (dados["Area"] <= 100)
response = dados[selecao].shape[0]
print(f"A quantidade de imóveis dentro de 60m e 100m é {response} unidades.")
```

Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00

```python
selecao = (dados["Quartos"] >= 4) & (dados["Valor"] < 2000)
response1 = dados[selecao].shape[0]
print(f'Á quantidade é {response1}')
```
